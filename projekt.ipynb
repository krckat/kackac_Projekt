```python
# KOCHOVA HVĚZDA(VLOČKA) v turtle

import turtle as tu

def t_kochova_hvezda(k, delka_strany):
    """Tato funkce vezme danou iteraci a delku strany a vykreslesli fraktal kochovy hvezdy pomoci knihovny turtle"""
    if k == 0:
        tu.forward(delka_strany)
    else:
        t_kochova_hvezda(k - 1, delka_strany)
        tu.left(60)
        t_kochova_hvezda(k - 1, delka_strany)
        tu.left(-120)
        t_kochova_hvezda(k - 1, delka_strany)
        tu.left(60) 
        t_kochova_hvezda(k - 1, delka_strany)   

tu.speed(0) # změně rychlosti.
tu.pencolor("mediumblue") # barva cary
tu.penup() # přestávky v kresleni.
tu.pensize(2) # tloustka cary.
delka_strany = 4
tu.backward(delka_strany / 2) # pohyb zpet.
tu.pendown() # kreslici poza.
for i in range(3):# dokresleni vsech 3 stran
    t_kochova_hvezda(4, delka_strany)
    tu.left(-120)

tu.done() # ukonceni.
```

```python
# kochova hvezda s matplotlib

import numpy as np
import matplotlib.pyplot as plt
import math as ma

def m_kochova_hvezda(k, delka_strany, pozice, uhel):
    """Tato funkce vezme danou iteraci,delku strany,uhela a pozici, pak se vykreslesli fraktal kochovy hvezdy pomoci knihovny matplotlib"""
    if k == 0: # usecka, vytvori se krajni body
        x = pozice[0] + delka_strany * np.cos(np.deg2rad(uhel))
        y = pozice[1] + delka_strany * np.sin(np.deg2rad(uhel))
        return [(pozice[0], pozice[1]), (x, y)] 
    
    else:
        vysledek = [] # list bodu
        vysledek += m_kochova_hvezda(k - 1, delka_strany, pozice, uhel)
        pozice = (vysledek[-1][0], vysledek[-1][1])
        uhel += 60
        vysledek += m_kochova_hvezda(k - 1, delka_strany, pozice, uhel)
        pozice = (vysledek[-1][0], vysledek[-1][1])
        uhel -= 120
        vysledek += m_kochova_hvezda(k - 1, delka_strany, pozice, uhel)
        pozice = (vysledek[-1][0], vysledek[-1][1])
        uhel += 60
        vysledek += m_kochova_hvezda(k - 1, delka_strany, pozice, uhel)
        return vysledek

def kochova_krivka(k, delka_strany):
    """tato funkce vezme delku strany a iterace a vzkresli dalsi body hvezdy"""
    pozice = (0, 0)
    uhel = 0
    vysledek = [(0, 0)]
    for i in range(3):
        vysledek += m_kochova_hvezda(k, delka_strany, pozice, uhel)
        pozice = vysledek[-1]
        uhel -= 120
    return vysledek

def plot_kochova_krivka(k, delka_strany):
    """tato funkce vzkresluje fraktal"""
    body = kochova_krivka(k, delka_strany)
    x = [p[0] for p in body]
    y = [p[1] for p in body]
    plt.plot(x, y)
    plt.axis('off')

plot_kochova_krivka(5, 1)
plt.show()
```

```python
# Strom v turtle   

import turtle as tu
tu.speed(0) # změně rychlosti. 
tu.pensize(2) # tloustka cary
tu.pencolor("saddlebrown") # barva cary

tu.left(90)

def t_strom(delka_strany, k):
    """tato funkce vezme delku strany a iterace a nasledne vykresli fraktal stromu v knihovne turtle"""
    if k > 0:
        tu.forward(delka_strany)
        tu.right(30)
        t_strom(0.8 * delka_strany, k - 1)
        tu.left(60)
        t_strom(0.8 * delka_strany, k - 1)
        tu.right(30)
        tu.backward(delka_strany)

t_strom(70, 6)     # velkost(70) iterace(6)
tu.done() # ukonceni.
```

```python
# strom v matplotlibu
import matplotlib.pyplot as plt

def m_strom(x, y, delka_strany, uhel, k):
    """tato funkce vezme delku strany a iterace a nasledne vykresli fraktal stromu v knihovne matplotlib"""
    if k > 0:
        # Nastaví nové souřadnice
        x2 = x + delka_strany * np.cos(np.deg2rad(uhel))
        y2 = y + delka_strany * np.sin(np.deg2rad(uhel))
        # rekurze pro vetve
        m_strom(x2, y2, 0.8 * delka_strany, uhel + 30, k - 1)
        m_strom(x2, y2, 0.8 * delka_strany, uhel - 60, k - 1)
        # Nakreslí
        plt.plot((x, x2), (y, y2), color='saddlebrown', linewidth=1)#linewidth sirka cary
        plt.axis('off')# vypnuti os
        
m_strom(0, 0, 50, 90, 5)
plt.show()
```

```python
# Pythagoruv strom v turtle

import turtle as tu
import math as ma

tu.speed(0) # změně rychlosti.
tu.pencolor("black") # barva cary
tu.pensize(2) # tloustka

def t_py_strom(delka_strany, k):
    """tato funkce vezme delku strany a iterace a nasledne vzkresli vetve v knihovne turtle"""
    if k == 0:
        tu.forward(delka_strany)
    else:
        tu.right(135)
        tu.forward(delka_strany * ma.sqrt(2)/2)
        tu.left(90)
        t_py_strom(delka_strany * ma.sqrt(2)/2, k - 1)
        tu.left(90)
        tu.forward(delka_strany * ma.sqrt(2)/2)
        tu.left(90)
        tu.forward(delka_strany * ma.sqrt(2)/2)
        tu.right(135)

        tu.forward(delka_strany)

        tu.right(135)
        tu.forward(delka_strany * ma.sqrt(2)/2)
        tu.left(90)
        tu.forward(delka_strany * ma.sqrt(2)/2)
        tu.left(90)
        t_py_strom(delka_strany * ma.sqrt(2)/2, k - 1)
        tu.left(90)
        tu.forward(delka_strany * ma.sqrt(2)/2)
        tu.right(135)

def ctverec(delka_strany, k):
    """tato funce vezme delku strany a iterace a nakresli kmen stromu v knihovne turtle"""
    tu.forward(delka_strany)
    tu.left(90)
    tu.forward(delka_strany)
    tu.left(90)
    t_py_strom(delka_strany, k)
    tu.left(90)
    tu.forward(delka_strany) 

ctverec(100, 3)

tu.done() # ukonceni.
```

```python
# Pythagoruv strom v matplotlibu
import numpy as np
import matplotlib.pyplot as plt
import math as ma

def m_py_strom(delka_strany, k, pozice=(0,0), uhel=0):
    """tato funkce vezme delku strany, iterace,pozici a uhel a nasledne vykresli vetve stromu pomoci knihovny matplotlib """
    if k == 0:
        x = pozice[0] + delka_strany * np.cos(np.deg2rad(uhel)) # vytvari usecku
        y = pozice[1] + delka_strany * np.sin(np.deg2rad(uhel))
        return [pozice, (x,y)]
    else:
        vysledek = [pozice]
        uhel -= 135
        delka = delka_strany * np.sqrt(2) / 2 #definice strany
        vysledek += [(pozice[0] + delka*np.cos(np.deg2rad(uhel)),pozice[1] + delka*np.sin(np.deg2rad(uhel)))]
        pozice = vysledek[-1] #kde jsme skoncili
        uhel += 90
        vysledek += m_py_strom(delka, k-1, pozice, uhel)
        uhel += 90
        pozice = vysledek[-1] # to dole dlani nove usecky
        vysledek += [(pozice[0] + delka*np.cos(np.deg2rad(uhel)),pozice[1] + delka*np.sin(np.deg2rad(uhel)))]
        uhel += 90
        pozice = vysledek[-1]
        vysledek += [(pozice[0] + delka*np.cos(np.deg2rad(uhel)),pozice[1] + delka*np.sin(np.deg2rad(uhel)))]
        uhel -= 135
        pozice = vysledek[-1]
        vysledek += [(pozice[0] + delka_strany*np.cos(np.deg2rad(uhel)),pozice[1] + delka_strany*np.sin(np.deg2rad(uhel)))]

        uhel -= 135
        pozice = vysledek[-1]
        vysledek += [(pozice[0] + delka*np.cos(np.deg2rad(uhel)),pozice[1] + delka*np.sin(np.deg2rad(uhel)))]
        uhel += 90
        pozice = vysledek[-1]
        vysledek += [(pozice[0] + delka*np.cos(np.deg2rad(uhel)),pozice[1] + delka*np.sin(np.deg2rad(uhel)))]
        uhel += 90
        pozice = vysledek[-1]
        vysledek += m_py_strom(delka, k-1, pozice, uhel)
        uhel += 90
        pozice = vysledek[-1]
        vysledek += [(pozice[0] + delka*np.cos(np.deg2rad(uhel)),pozice[1] + delka*np.sin(np.deg2rad(uhel)))]
        return vysledek

def ctverec(delka_strany, k, pozice=(0,0), uhel=0):
    """tato funkce vezme delku strany,iterace,pozici a uhel a vykresli kmen stromu v knihovne matplotlib  """
    vysledek = [pozice]# list bodu ctverce
    A = (pozice[0] + delka_strany*np.cos(np.deg2rad(uhel)), pozice[1] + delka_strany*np.sin(np.deg2rad(uhel)))
    B = (A[0] + delka_strany*np.cos(np.deg2rad(uhel+90)), A[1] + delka_strany*np.sin(np.deg2rad(uhel+90)))
    vysledek += [A,B]#vrcholy ctverce
    vysledek += m_py_strom(delka_strany, k, B, uhel+180)
    vysledek += [pozice]# dokresleni posledni cary
    return vysledek
    
def plot_py_strom(delka_strany, k):
    """tato funkce vykresluje fraktal"""
    body = ctverec(delka_strany, k)
    x = [p[0] for p in body]
    y = [p[1] for p in body]
    plt.plot(x, y, 'k')
    plt.axis('off')

plot_py_strom(1, 5)
plt.show()
```
